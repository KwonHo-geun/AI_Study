```
import time
import io
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image, ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True

from rembg.bg import remove #remove background

def show_rembg(path):
    fig = plt.figure(figsize=(10, 10))

    # 원본 이미지 출력
    fig.add_subplot(1, 2, 1)
    orig_img = Image.open(path)
    plt.imshow(orig_img)
    plt.axis('off')

    # 배경 제거
    fig.add_subplot(1, 2, 2)

    with open(path, "rb") as f:
        input_data = f.read()  # 이미지 파일을 바이너리로 읽기

    started = time.time()
    result = remove(input_data)
    elapsed = time.time() - started
    print(f'it takes {elapsed} seconds for removing bg.')

    img = Image.open(io.BytesIO(result)).convert("RGBA")
    plt.imshow(img)
    plt.axis('off')
```

```
car_path = "/content/car.jpg"
ele_path = "/content/elephant.jpg"
tiger_path = "/content/tiger.jpg"
```

```
show_rembg(tiger_path)
```

```
show_rembg(car_path)
```
