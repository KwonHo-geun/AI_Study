```
!pip install ultralytics yt-dlp opencv-python

import cv2
import os
from ultralytics import YOLO
from IPython.display import Video, display

def download_and_analyze_youtube_video(youtube_url):
    """
    YouTube ì˜ìƒì„ ì§ì ‘ ë‹¤ìš´ë¡œë“œí•˜ê³  YOLO ë¶„ì„
    """
    print("â¬‡ï¸ YouTube ì˜ìƒ ë‹¤ìš´ë¡œë“œ ì¤‘...")

    # YouTube ì˜ìƒ ë‹¤ìš´ë¡œë“œ (yt-dlp ì§ì ‘ ëª…ë ¹ì–´ ì‚¬ìš©)
    os.system(f'yt-dlp -f "bestvideo+bestaudio/best[height<=720]" --merge-output-format mp4 "{youtube_url}"')

    # ë‹¤ìš´ë¡œë“œëœ mp4 íŒŒì¼ ì°¾ê¸°
    video_file = None
    for file in os.listdir("/content"):
        if file.endswith(".mp4"):
            video_file = file
            print(f"ğŸ“ ë‹¤ìš´ë¡œë“œëœ ì˜ìƒ: {file}")
            break

    if not video_file:
        print("âŒ ë‹¤ìš´ë¡œë“œëœ ì˜ìƒì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        return None

    # YOLO ëª¨ë¸ ë¡œë“œ
    print("ğŸ¤– YOLO ëª¨ë¸ ë¡œë“œ ì¤‘...")
    model = YOLO("yolov8n.pt")

    # ë¹„ë””ì˜¤ ìº¡ì²˜ ì„¤ì •
    print("ğŸ¬ ì˜ìƒ ë¶„ì„ ì‹œì‘...")
    video_path = f"/content/{video_file}"
    cap = cv2.VideoCapture(video_path)

    # ë¹„ë””ì˜¤ ì •ë³´ ê°€ì ¸ì˜¤ê¸°
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    duration = total_frames / fps

    print(f"ğŸ“¹ ì˜ìƒ ì •ë³´:")
    print(f"   í•´ìƒë„: {width}x{height}")
    print(f"   FPS: {fps:.1f}")
    print(f"   ì´ í”„ë ˆì„: {total_frames}")
    print(f"   ê¸¸ì´: {duration:.1f}ì´ˆ")

    # ê²°ê³¼ ì €ì¥ ì„¤ì •
    output_path = "result_video_youtube.mp4"
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

    # ë¶„ì„ í†µê³„
    frame_count = 0
    detection_count = 0
    frames_with_objects = 0

    # í”„ë ˆì„ë³„ë¡œ ì¶”ë¡  ë° ì €ì¥
    print("ğŸ” YOLO ë¶„ì„ ì§„í–‰ ì¤‘...")

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # YOLOv8 ì¶”ë¡ 
        results = model(frame, verbose=False, conf=0.5)

        # íƒì§€ëœ ê°ì²´ ìˆ˜ ê³„ì‚°
        if len(results) > 0 and results[0].boxes is not None:
            num_detections = len(results[0].boxes)
            detection_count += num_detections
            if num_detections > 0:
                frames_with_objects += 1

        # ê²°ê³¼ ì´ë¯¸ì§€ (bounding box í¬í•¨)
        annotated_frame = results[0].plot()
        out.write(annotated_frame)

        frame_count += 1

        # ì§„í–‰ë¥  í‘œì‹œ (10% ë‹¨ìœ„)
        if frame_count % (total_frames // 10) == 0:
            progress = (frame_count / total_frames) * 100
            print(f"ì§„í–‰ë¥ : {progress:.0f}%")

    # ë¦¬ì†ŒìŠ¤ í•´ì œ
    cap.release()
    out.release()

    # ë¶„ì„ ê²°ê³¼ í†µê³„
    print("\n" + "="*50)
    print("ğŸ“Š ë¶„ì„ ì™„ë£Œ! ê²°ê³¼ í†µê³„")
    print("="*50)
    print(f"ì´ ì²˜ë¦¬ í”„ë ˆì„: {frame_count}")
    print(f"ì´ íƒì§€ëœ ê°ì²´: {detection_count}")
    print(f"ê°ì²´ê°€ íƒì§€ëœ í”„ë ˆì„: {frames_with_objects}")
    if frame_count > 0:
        print(f"í”„ë ˆì„ë‹¹ í‰ê·  íƒì§€ ìˆ˜: {detection_count/frame_count:.2f}")
        print(f"ê°ì²´ íƒì§€ìœ¨: {(frames_with_objects/frame_count)*100:.1f}%")

    print(f"\nâœ… ê²°ê³¼ ì˜ìƒ ì €ì¥: {output_path}")

    # ê²°ê³¼ ì˜ìƒ í‘œì‹œ
    print("ğŸ¥ ê²°ê³¼ ì˜ìƒ ì¬ìƒ:")
    return output_path

def analyze_local_video(video_path):
    """
    ë¡œì»¬ ë¹„ë””ì˜¤ íŒŒì¼ YOLO ë¶„ì„ (ì—…ë¡œë“œëœ íŒŒì¼ìš©)
    """
    if not os.path.exists(video_path):
        print(f"âŒ íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {video_path}")
        return None

    # YOLO ëª¨ë¸ ë¡œë“œ
    print("ğŸ¤– YOLO ëª¨ë¸ ë¡œë“œ ì¤‘...")
    model = YOLO("yolov8n.pt")

    # ë¹„ë””ì˜¤ ìº¡ì²˜ ì„¤ì •
    print("ğŸ¬ ì˜ìƒ ë¶„ì„ ì‹œì‘...")
    cap = cv2.VideoCapture(video_path)

    # ë¹„ë””ì˜¤ ì •ë³´ ê°€ì ¸ì˜¤ê¸°
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    duration = total_frames / fps if fps > 0 else 0

    print(f"ğŸ“¹ ì˜ìƒ ì •ë³´:")
    print(f"   í•´ìƒë„: {width}x{height}")
    print(f"   FPS: {fps:.1f}")
    print(f"   ì´ í”„ë ˆì„: {total_frames}")
    print(f"   ê¸¸ì´: {duration:.1f}ì´ˆ")

    # ê²°ê³¼ ì €ì¥ ì„¤ì •
    output_path = "result_video_local.mp4"
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

    # ë¶„ì„ í†µê³„
    frame_count = 0
    detection_count = 0
    frames_with_objects = 0

    # í”„ë ˆì„ë³„ë¡œ ì¶”ë¡  ë° ì €ì¥
    print("ğŸ” YOLO ë¶„ì„ ì§„í–‰ ì¤‘...")

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # YOLOv8 ì¶”ë¡ 
        results = model(frame, verbose=False, conf=0.5)

        # íƒì§€ëœ ê°ì²´ ìˆ˜ ê³„ì‚°
        if len(results) > 0 and results[0].boxes is not None:
            num_detections = len(results[0].boxes)
            detection_count += num_detections
            if num_detections > 0:
                frames_with_objects += 1

        # ê²°ê³¼ ì´ë¯¸ì§€ (bounding box í¬í•¨)
        annotated_frame = results[0].plot()
        out.write(annotated_frame)

        frame_count += 1

        # ì§„í–‰ë¥  í‘œì‹œ (10% ë‹¨ìœ„)
        if total_frames > 0 and frame_count % max(1, total_frames // 10) == 0:
            progress = (frame_count / total_frames) * 100
            print(f"ì§„í–‰ë¥ : {progress:.0f}%")

    # ë¦¬ì†ŒìŠ¤ í•´ì œ
    cap.release()
    out.release()

    # ë¶„ì„ ê²°ê³¼ í†µê³„
    print("\n" + "="*50)
    print("ğŸ“Š ë¶„ì„ ì™„ë£Œ! ê²°ê³¼ í†µê³„")
    print("="*50)
    print(f"ì´ ì²˜ë¦¬ í”„ë ˆì„: {frame_count}")
    print(f"ì´ íƒì§€ëœ ê°ì²´: {detection_count}")
    print(f"ê°ì²´ê°€ íƒì§€ëœ í”„ë ˆì„: {frames_with_objects}")
    if frame_count > 0:
        print(f"í”„ë ˆì„ë‹¹ í‰ê·  íƒì§€ ìˆ˜: {detection_count/frame_count:.2f}")
        print(f"ê°ì²´ íƒì§€ìœ¨: {(frames_with_objects/frame_count)*100:.1f}%")

    print(f"\nâœ… ê²°ê³¼ ì˜ìƒ ì €ì¥: {output_path}")

    return output_path

# ë©”ì¸ ì‹¤í–‰ ë¶€ë¶„
print("ğŸ¬ YouTube YOLO ë¶„ì„ê¸° (ê°„ë‹¨ ë²„ì „)")
print("="*50)

mode = input("\nì„ íƒí•˜ì„¸ìš”:\n1. YouTube URL ì…ë ¥\n2. íŒŒì¼ ì—…ë¡œë“œ\n\nì„ íƒ (1 ë˜ëŠ” 2): ")

if mode == "1":
    # YouTube URL ì…ë ¥
    youtube_url = input("\nYouTube URLì„ ì…ë ¥í•˜ì„¸ìš”: ")

    result_path = download_and_analyze_youtube_video(youtube_url)

    if result_path and os.path.exists(result_path):
        # ê²°ê³¼ ì˜ìƒ í‘œì‹œ
        display(Video(result_path, embed=True, width=800))
    else:
        print("âŒ ë¶„ì„ ì‹¤íŒ¨")

elif mode == "2":
    # íŒŒì¼ ì—…ë¡œë“œ
    print("\nğŸ“ ë¹„ë””ì˜¤ íŒŒì¼ì„ ì—…ë¡œë“œí•˜ì„¸ìš”:")
    try:
        from google.colab import files
        uploaded = files.upload()

        if uploaded:
            video_file = list(uploaded.keys())[0]
            print(f"ğŸ“¹ ì—…ë¡œë“œëœ íŒŒì¼: {video_file}")

            result_path = analyze_local_video(video_file)

            if result_path and os.path.exists(result_path):
                # ê²°ê³¼ ì˜ìƒ í‘œì‹œ
                display(Video(result_path, embed=True, width=800))
            else:
                print("âŒ ë¶„ì„ ì‹¤íŒ¨")
        else:
            print("âŒ íŒŒì¼ì´ ì—…ë¡œë“œë˜ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤.")

    except ImportError:
        # Colabì´ ì•„ë‹Œ í™˜ê²½
        video_path = input("ë¹„ë””ì˜¤ íŒŒì¼ ê²½ë¡œë¥¼ ì…ë ¥í•˜ì„¸ìš”: ")
        result_path = analyze_local_video(video_path)

        if result_path and os.path.exists(result_path):
            print(f"âœ… ê²°ê³¼ ì˜ìƒ: {result_path}")
        else:
            print("âŒ ë¶„ì„ ì‹¤íŒ¨")

else:
    print("âŒ ì˜ëª»ëœ ì„ íƒì…ë‹ˆë‹¤.")

print("\nğŸ‰ ì‘ì—… ì™„ë£Œ!")
```
